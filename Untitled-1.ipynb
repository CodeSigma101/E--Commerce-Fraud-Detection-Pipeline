```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Load the transactional dataset
df = pd.read_csv('Dataset for Data Analytics - Sheet1.csv')


# 2. Define the fraudulent target label (e.g., Cancelled orders as a proxy for fraudulent chargebacks)
df['is_fraud'] = df['OrderStatus'].apply(lambda x: 1 if x == 'Cancelled' else 0)

# 3. Verify baseline metrics
print("--- Dataset Summary ---")
print(f"Total Transactions: {df.shape[0]}")
print(f"Total Columns: {df.shape[1]}\n")

# 4. Analyze target class distribution
class_counts = df['is_fraud'].value_counts()
class_percentages = df['is_fraud'].value_counts(normalize=True) * 100

print("--- Class Distribution ---")
print(f"Legitimate Transactions (0): {class_counts[0]} ({class_percentages[0]:.2f}%)")
print(f"Fraudulent Transactions (1): {class_counts[1]} ({class_percentages[1]:.2f}%)\n")

# 5. Visualize the Imbalance
plt.figure(figsize=(6, 4))
sns.countplot(x='is_fraud', data=df, palette='viridis')
plt.title('Severe Transaction Class Imbalance')
plt.xlabel('Is Fraud? (0 = No, 1 = Yes)')
plt.ylabel('Transaction Count')
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from imblearn.over_sampling import SMOTE

# 1. Feature Engineering: Convert Date to useful numerical features
df['Date'] = pd.to_datetime(df['Date'])
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Day'] = df['Date'].dt.day

# 2. Drop high-cardinality metadata columns that cause noise
cols_to_drop = ['OrderID', 'CustomerID', 'TrackingNumber', 'ShippingAddress', 'Date', 'OrderStatus', 'is_fraud']
X = df.drop(columns=cols_to_drop)
y = df['is_fraud']

# 3. Handle categorical values using One-Hot Encoding
X = pd.get_dummies(X, columns=['Product', 'PaymentMethod', 'CouponCode', 'ReferralSource'], drop_first=True)

# 4. Stratified Train-Test Split (Crucial step BEFORE SMOTE to prevent data leakage)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, stratify=y, random_state=42)

# 5. Scale the numerical value distributions
scaler = StandardScaler()
numerical_cols = ['Quantity', 'UnitPrice', 'ItemsInCart', 'TotalPrice', 'Year', 'Month', 'Day']
X_train[numerical_cols] = scaler.fit_transform(X_train[numerical_cols])
X_test[numerical_cols] = scaler.transform(X_test[numerical_cols])

# 6. Apply SMOTE to training partition only
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train, y_train)

print("--- Data Splitting & Balance Verification ---")
print(f"Original Training Size: {X_train.shape[0]} samples")
print(f"Resampled Training Size (SMOTE): {X_train_resampled.shape[0]} samples")
print(f"Test Partition Size: {X_test.shape[0]} samples\n")
print("Class Distribution After SMOTE:")
print(y_train_resampled.value_counts())
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier

# 1. Initialize the machine learning algorithms
log_reg = LogisticRegression(max_iter=1000, random_state=42)
rf_clf = RandomForestClassifier(n_estimators=100, random_state=42)

print("Training Logistic Regression model...")
# 2. Fit Logistic Regression on resampled training data
log_reg.fit(X_train_resampled, y_train_resampled)

print("Training Random Forest model...")
# 3. Fit Random Forest on resampled training data
rf_clf.fit(X_train_resampled, y_train_resampled)

# 4. Generate predictions using the original, un-sampled test split
y_pred_lr = log_reg.predict(X_test)
y_pred_rf = rf_clf.predict(X_test)

# Generate probability scores (needed later for ROC-AUC metrics)
y_prob_lr = log_reg.predict_proba(X_test)[:, 1]
y_prob_rf = rf_clf.predict_proba(X_test)[:, 1]

print("\nBoth models trained successfully on balanced data!")
```

```python
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score
import seaborn as sns
import matplotlib.pyplot as plt

def evaluate_model(name, y_true, y_pred, y_prob):
    print(f"\n=================== {name} Performance ===================")
    # 1. Print standard precision, recall, and f1-score summary
    print(classification_report(y_true, y_pred))
    
    # 2. Calculate the strict ROC-AUC metric
    roc_auc = roc_auc_score(y_true, y_prob)
    print(f"ROC-AUC Score: {roc_auc:.4f}")
    
    # 3. Visualize the Confusion Matrix
    cm = confusion_matrix(y_true, y_pred)
    plt.figure(figsize=(5, 3.5))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', 
                xticklabels=['Legit (0)', 'Fraud (1)'], 
                yticklabels=['Legit (0)', 'Fraud (1)'])
    plt.title(f'{name} Confusion Matrix')
    plt.ylabel('Actual Label')
    plt.xlabel('Predicted Label')
    plt.show()

# Run strict metric evaluations for both algorithms
evaluate_model("Logistic Regression", y_test, y_pred_lr, y_prob_lr)
evaluate_model("Random Forest", y_test, y_pred_rf, y_prob_rf)
```
